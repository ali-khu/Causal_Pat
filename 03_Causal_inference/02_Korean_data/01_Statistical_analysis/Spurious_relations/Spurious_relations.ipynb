```python
import pandas as pd
import numpy as np
import seaborn as sns


import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_excel('data.xlsx')
```

```python
x = df.copy()
x = x.loc[(x['head_house_edu_group'] != -2) & (x['HDD18'] != -2)]
```

```python
plt.figure(figsize=(8,5))
custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="ticks", rc=custom_params, font_scale=1.5)
plt.rcParams["axes.edgecolor"] = "black"
draw = sns.boxplot(x = x['head_house_edu_group'], y = x['HDD18'], width=0.4)
draw.set_xlabel("Income category",fontsize=24)
draw.set_ylabel("HDD [Degree days]",fontsize=24)
# draw.set(ylim=(0, 25000))
figure = draw.get_figure()
figure.savefig('Year_HDD.png', bbox_inches='tight', dpi=300)
```

```python
x = df.copy()
x = x.loc[(x['av_monthly_income_group'] != -2) & (x['HDD18'] != -2)]

plt.figure(figsize=(8,5))
custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="ticks", rc=custom_params, font_scale=1.5)
plt.rcParams["axes.edgecolor"] = "black"
draw = sns.boxplot(x = x['av_monthly_income_group'], y = x['HDD18'], width=0.4)
draw.set_xlabel("Income category",fontsize=24)
draw.set_ylabel("HDD [Degree days]",fontsize=24)
# draw.set(ylim=(0, 25000))
figure = draw.get_figure()
figure.savefig('Year_HDD.png', bbox_inches='tight', dpi=300)
```

```python
x = df.copy()
x = x.loc[(x['housing_direction'] != -2) & (x['res_area'] != -2)]

x["housing_direction"] = x['housing_direction'].replace([1, 2, 3, 4, 5, 6, 7, 8], ["East", "West", "South", "North", "South East", "South West", "North East", "North West"])

plt.figure(figsize=(15,5))
custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="ticks", rc=custom_params, font_scale=1.5)
plt.rcParams["axes.edgecolor"] = "black"
draw = sns.boxplot(x = x['housing_direction'], y = x['res_area'], width=0.4, order=["East", "West", "South", "North", "South East", "South West", "North East", "North West"])
draw.set_xlabel("Housing direction",fontsize=24)
draw.set_ylabel("Floor area ($\mathregular{m^2}$)",fontsize=24)
# draw.set(ylim=(0, 25000))
figure = draw.get_figure()
figure.savefig('Year_HDD.png', bbox_inches='tight', dpi=300)
```
